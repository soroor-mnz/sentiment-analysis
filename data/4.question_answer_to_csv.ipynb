```python
df_1 = pd.read_excel('/content/drive/My Drive/thesis/data/1.student-feedback.xlsx')
df_2 = pd.read_excel('/content/drive/My Drive/thesis/data/2.student-feedback.xlsx')
df_3 = pd.read_excel('/content/drive/My Drive/thesis/data/3.student-feedback.xlsx')
```

remove extra columns

```python
column_names = df_1.columns
extra_column = column_names[0]
df_1 = df_1.drop([extra_column], axis=1)

column_names_2 = df_2.columns
for cols in range(6):
  df_2 = df_2.drop([column_names_2[cols]], axis=1)

column_names_3 = df_3.columns
for cols in range(6):
  df_3 = df_3.drop([column_names_3[cols]], axis=1)
```

add questions and answers from all df_1 columns to 2D array

```python
column_names = df_1.columns

values = []
for col in column_names:
  for row in df_1[col].tolist():
    if type(row) is str :
      row = row.replace("اما" , ".")
      row = row.replace("ولی", ".")
      for sentence in row.split("."):
        values.append([sentence, col])

column_names_2 = df_2.columns

for col in column_names_2:
  for row in df_2[col].tolist():
    if type(row) is str :
      row = row.replace("اما" , ".")
      row = row.replace("ولی", ".")
      for sentence in row.split("."):
        values.append([sentence, col])

column_names_3 = df_3.columns

for col in column_names_3:
  for row in df_3[col].tolist():
    if type(row) is str :
      row = row.replace("اما" , ".")
      row = row.replace("ولی", ".")
      for sentence in row.split("."):
        values.append([sentence, col])
```

```python
dataframe = pd.DataFrame(values)
dataframe.columns = ["answer" , "question"]
```

```python
dataframe.to_csv(r'/content/drive/My Drive/thesis/data/5.merged_data.csv')
```
